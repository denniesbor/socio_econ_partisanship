```python
# congressional district personal and demographic information
df_house = pd.read_csv(os.path.join(WORKING_DIR, "117th Congress Members Guide with Election Results and Demographic Data by District - House.csv"))
prev_header = ""
columns = []
for i, column in enumerate(df_house.columns):
    if "Unnamed" not in column:
        prev_header = column
        subheader = df_house.iloc[0, i]
        if not pd.isna(subheader):
            column = f"{column}_{subheader}"
    else:
        subheader = df_house.iloc[0, i]
        if not pd.isna(subheader):
            column = f"{prev_header}_{subheader}"

    columns.append(column)


df_house.columns = columns
df_house.drop([0,1], inplace=True)
```

```python
# Cleaning
df_house["District"] = df_house["Code"].apply(
    lambda code: 0 if code.split("-")[1] == "AL" else int(code.split("-")[1])
)
df_house["State"] = df_house["Code"].apply(lambda code: code.split("-")[0])
df_house["Gender"] = df_house["Gender"].apply(
    lambda gender: "M" if gender == "Man" else "F" if gender == "Woman" else gender
)

# rename the columns
rename_mapping = {
    "Party": "Party",
    'First Elected': 'FirstElected',
    'Birth Year': 'BirthYear',
    'Gender': 'Gender',
    'Race/ Ethnicity': 'RaceEthnicity',
    'Religion': 'Religion',
    'LGBTQ': 'LGBTQ',
    'Pre-2020 Incumbent': 'Pre2020Incumbent',
    'Incumbent Name_First': 'FirstName',
    'Incumbent Name_Last': 'LastName',
    '2015-2019 ACS Citizen Adult Population_White': 'ACS_AdultPop_White',
    '2015-2019 ACS Citizen Adult Population_Black': 'ACS_AdultPop_Black',
    '2015-2019 ACS Citizen Adult Population_Latino': 'ACS_AdultPop_Latino',
    '2015-2019 ACS Citizen Adult Population_Asian and Pacific Islander': 'ACS_AdultPop_Asian_PI',
    '2015-2019 ACS Citizen Adult Population_Native': 'ACS_AdultPop_Native',
    '2015-2019 ACS Citizen Adult Population_Other': 'ACS_AdultPop_Other',
    "2015-2019 ACS Educational Attainment Among Adults 25+ and Median Household Income_Bachelor's Degree or Higher": 'ACS_EdAttainment_BachelorsPlus',
    '2015-2019 ACS Educational Attainment Among Adults 25+ and Median Household Income_Median Income': 'ACS_MedianIncome',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_Total': 'ACS_Pop_Total',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_White': 'ACS_Pop_White',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_Black': 'ACS_Pop_Black',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_Latino': 'ACS_Pop_Latino',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_Asian and Pacific Islander': 'ACS_Pop_Asian_PI',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_Native American': 'ACS_Pop_NativeAmerican',
    '2015-2019 ACS Citizen Population (Includes Latinos of Any Race)_Other': 'ACS_Pop_Other',
    '2015-2019 ACS Educational Attainment_Total': 'ACS_EdAttainment_Total',
    '2015-2019 ACS Educational Attainment_College': 'ACS_EdAttainment_College',
    'District': 'DistrictAbbrev',
    'State': 'State'
}

df_house_remapped = df_house[list(rename_mapping.keys())].rename(columns=rename_mapping)

combine_clean_columns = {
    'ACS_AdultPop_Others': ['ACS_AdultPop_Asian_PI', 'ACS_AdultPop_Native', 'ACS_AdultPop_Other'],
    'ACS_POP_Others': ['ACS_Pop_Asian_PI', 'ACS_Pop_NativeAmerican', 'ACS_Pop_Other'],
}
for new_col, cols in combine_clean_columns.items():
    df_house_remapped[new_col] = df_house_remapped[cols].apply(
        lambda x: x.str.replace(',', '').astype(float).sum(), axis=1
    )

df_house_remapped.drop(columns=sum(combine_clean_columns.values(), []), axis=1, inplace=True)

# Cleaning and dividing ACS Education Attainment columns
df_house_remapped['ACS_EdAttainmentCol'] = (
    df_house_remapped['ACS_EdAttainment_College'].str.replace(',', '').astype(float) /
    df_house_remapped['ACS_EdAttainment_Total'].str.replace(',', '').astype(float)
)
df_house_remapped.drop(columns=['ACS_EdAttainment_College', 'ACS_EdAttainment_Total'], axis=1, inplace=True)

#  clean median income
df_house_remapped["ACS_MedianIncome"] = (
    df_house_remapped["ACS_MedianIncome"]
    .replace({r"\$": "", ",": "", " ": ""}, regex=True)
    .apply(pd.to_numeric, errors='coerce')
)
median_income = df_house_remapped["ACS_MedianIncome"].median()
df_house_remapped["ACS_MedianIncome"].fillna(median_income, inplace=True)

# more columns to clean
columns_to_clean = [
    'ACS_AdultPop_White',
    'ACS_AdultPop_Black',
    'ACS_AdultPop_Latino',
    'ACS_EdAttainment_BachelorsPlus',
    'ACS_Pop_Total',
    'ACS_Pop_White',
    'ACS_Pop_Black',
    'ACS_Pop_Latino'
]

for col in columns_to_clean:
    df_house_remapped[col] = df_house_remapped[col].str.replace(',', '').astype(float)

# get percentage of population by race
pop_cols = ['ACS_Pop_White','ACS_Pop_Black','ACS_Pop_Latino', 'ACS_POP_Others']

for col in pop_cols:
  df_house_remapped[col] = ((df_house_remapped[col] / df_house_remapped["ACS_Pop_Total"]) * 100).round(2)

df_house_remapped.drop("ACS_Pop_Total", axis=1, inplace=True)

df = df_house_remapped[['FirstName',
 'LastName',
  'Party',
  'DistrictAbbrev',
 'State',
 'FirstElected',
 'BirthYear',
 'Gender',
 'RaceEthnicity',
 'Religion',
 'LGBTQ',
 'Pre2020Incumbent',
 'ACS_AdultPop_White',
 'ACS_AdultPop_Black',
 'ACS_AdultPop_Latino',
'ACS_AdultPop_Others',
 'ACS_EdAttainment_BachelorsPlus',
 'ACS_EdAttainmentCol',
 'ACS_MedianIncome',
 'ACS_Pop_White',
 'ACS_Pop_Black',
 'ACS_Pop_Latino',
  'ACS_POP_Others']]

df.BirthYear = df.BirthYear.apply(pd.to_numeric, errors="coerce")

# fill na with median age of birth
median_year = df["BirthYear"].median()
df["BirthYear"].fillna(median_year, inplace=True)
df.BirthYear = df.BirthYear.astype(int).astype(str)

# # Age
today = datetime.date.today()
df["BirthYear"] = pd.to_datetime(df["BirthYear"])
df["Age"] = df["BirthYear"].apply(
    lambda x: today.year - x.year
)

df.drop("BirthYear", axis=1, inplace=True)
```

```python
def get_veterans(url):
    """
    Scrape the veterans data from the given URL using Selenium.
    """
    # Create a new Chrome session
    driver = webdriver.Chrome()
    wait = WebDriverWait(driver, 20)

    # Load the web page and maximize window
    driver.get(url)
    driver.maximize_window()

    # Wait for the required element to be visible
    wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".igc-table-container")))

    # Short pause for complete loading
    time.sleep(0.5)

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, "html.parser")
    div = soup.select_one("div.igc-table-container")
    df = pd.read_html(str(div))[0]

    # Data cleaning and manipulation
    df['State'] = df['District'].apply(
        lambda dis: dis.split(" ")[0] if dis != "Georgia" else "GA"
    )
    df['District'] = df['District'].apply(
        lambda dis: "0" if len(dis.split(" ")) == 1 or dis.split(" ")[1] == "(House)" else str(dis.split(" ")[1])
    )
    df['DistrictAbbrev'] = df['District'].apply(
        lambda dis: "Senator" if dis == "Sen" or dis == "Senate" else dis
    )
    df['Party'] = df['Party'].apply(
        lambda party: "R" if party == "GOP" or party == "Republican" else "D" if party == "Dem" else party
    )

    # Drop unnecessary columns and filter by 'House'
    df.drop(columns=['District'], axis=1, inplace=True)
    return df[df['Chamber'] == 'House']

# Read or scrape veterans data
file_path = os.path.join(WORKING_DIR, "veterans_in_congress_117.csv")
if not os.path.exists(file_path):
    df_mil_2022 = get_veterans("https://infogram.com/veterans-in-117th-congress-1hxj48ppv9zz52v")
    df_mil_2022.to_csv(file_path, index=False)

df_military = pd.read_csv(file_path)
print(df_military.head(5))
```

```python
# Merge df with df_vets using name and state
df_vets_house_names = df_military['Name'].str.split(expand=True)

# Check if either first or last name from df matches names in df_vets_house_names
df['is_in_df2'] = (
    df['FirstName'].isin(df_vets_house_names[0]) |
    df['LastName'].isin(df_vets_house_names[0]) |
    df['FirstName'].isin(df_vets_house_names[1]) |
    df['LastName'].isin(df_vets_house_names[1])
)

# Filter rows where the name matches
is_vet_house = df[df['is_in_df2']]

# Clean up by dropping the 'is_in_df2' column
df.drop('is_in_df2', axis=1, inplace=True)

# Merge with df_military to identify veterans
merged_df_vet_house = df_military.merge(
    is_vet_house[['FirstName', 'LastName', 'DistrictAbbrev', 'State']],
    on=['DistrictAbbrev', 'State'],
    how='left'
)
merged_df_vet_house['IsVet'] = 'Yes'

# Final merge to add veteran status
df_vet = df.merge(
    merged_df_vet_house[['State', 'DistrictAbbrev', 'FirstName', 'IsVet']],
    on=['State', 'DistrictAbbrev', 'FirstName'],
    how='left'
)
df_vet['IsVet'] = df_vet['IsVet'].fillna('No')

# Load and preview ideology data
df_ideology = pd.read_csv(os.path.join(WORKING_DIR, "govtrack-stats-2022-house-ideology.csv"))
print(df_ideology.head(5))

# Rename columns for consistency
df_ideology.rename(
    columns={'district': 'DistrictAbbrev', 'id': 'govtrack_d', 'state': 'State'},
    inplace=True
)
```

```python
def decode_byte_string(s):
    try:
        # Convert the literal byte string to an actual byte string
        byte_str = ast.literal_eval(s)
        # Decode the byte string
        return byte_str.decode('utf-8')
    except (ValueError, SyntaxError):
        # Return the string as is if it's not a byte string
        return s

df_ideology['name'] = df_ideology['name'].apply(decode_byte_string)
```

### Geographically Weighted Regression

```python
y = region_gdf["ideology"]

X = ["ACS_EdAttainment_BachelorsPlus", "ACS_Pop_White", "ACS_MedianIncome"]
g_X = region_gdf[X].values
g_y = y.values.reshape((-1,1))

# calculate centroid
u = region_gdf.geometry.centroid.x.values
v = region_gdf.geometry.centroid.y.values

g_coords = list(zip(u,v))

# compute z score of X
g_X = (g_X - g_X.mean(axis=0)) / g_X.std(axis=0)

#  standardixe the y values
g_y = (g_y - g_y.mean(axis=0)) / g_y.std(axis=0)

# select bandwidth of neighbors
gwr_selector = Sel_BW(g_coords, g_y, g_X)

gwr_bw = gwr_selector.search(bw_min=2)
print(gwr_bw)

# fit the model
gwr_results = GWR(g_coords, g_y, g_X, gwr_bw).fit()
```
